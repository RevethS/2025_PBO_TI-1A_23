```python
#Praktikum 1
class Burung:
  def __init__(self, nama):
    self.nama = nama

  def terbang(self):
    print(f"{self.nama} terbang dengan cara umum.")

  def bersuara(self):
    print(f"{self.nama} mengeluarkan suara burung.")

class Elang(Burung):
  def __init__(self, nama, rentang_sayap):
    super().__init__(nama)
    self.rentang_sayap = rentang_sayap

  def terbang(self):
    print(f"{self.nama} terbang tinggi melayang diangkasa.")
  def bersuara(self):
    print(f"{self.nama} berteriak nyaring!")

class Pipit(Burung):
  def __init__(self, nama, warna_bulu):
    super().__init__(nama)
    self.warna_bulu = warna_bulu

  def terbang(self):
    print(f"{self.nama} terbang cepat diantara pepohonan.")

  def bersuara(self):
    print(f"{self.nama} berkicau merdu: Cit Cit!")

def demonstrasi_aksi_burung(daftar_burung):
  print("\nAksi Burung:")

  for burung in daftar_burung:
      print(f"-- Aksi untuk {burung.nama} --")
      burung.terbang()
      burung.bersuara()
      print("-" * 40)
if __name__ == "__main__":
  elang_jawa = Elang("Elang Jawa", 1.5)
  pipit_gereja = Pipit("Pipit Gereja", "Coklat")
  burung_aneh = Burung("Burung Mysterius")

  daftar_burung = [elang_jawa, pipit_gereja, burung_aneh]
  demonstrasi_aksi_burung(daftar_burung)
```

```python
class Kucing:
  def bersuara(self):
    print("Kucing: Meow!")
class Anjing:
  def bersuara(self):
    print("Anjing: Dawg!")
class Bebek:
  def bersuara(self):
    print("Bebek: Kwek kwek!")
class Mobil:
  def jalanan(self):
    print("Mobil: Brumm!")

def buat_suara(objek_yang_bisa_bersuara):
  try:
    objek_yang_bisa_bersuara.bersuara()
  except AttributeError:
    print(f"{type(objek_yang_bisa_bersuara).__name__} tidak bisa bersuara.")
if __name__ == "__main__":
  kucing1 = Kucing()
  anjing1 = Anjing()
  bebek1 = Bebek()
  mobil1 = Mobil()

  daftar_objek = [kucing1, anjing1, bebek1, mobil1]
  print("Demonstrasi Duck Typing:")

  for item in daftar_objek:
    buat_suara(item)
    print("-" * 40)
```

```python
class KalkulatorSederhana:
    def __init__(self, nama="Kalkulator"):
        self.nama = nama
        print(f"{self.nama} siap digunakan.")

    def tambah(self, *args):
        print(f"\nMemanggil metode tambah dengan argumen: {args}")
        if not args:
            print("Tidak ada angka untuk dijumlahkan.")
            return 0

        total = 0
        valid_input = True

        for arg in args:
            if isinstance(arg, (int, float)):
                total += arg
            else:
                print(f"Input tidak valid dan diabaikan: {arg}")
                valid_input = False

        if valid_input:
            print(f"Hasil Penjumlahan: {total}")
        else:
            print(f"Hasil Penjumlahan (dengan beberapa input diabaikan): {total}")
        return total

if __name__ == "__main__":
    kalkulator = KalkulatorSederhana("Calc-01")
    print("\n---Percobaan Penjumlahan---")
    kalkulator.tambah(5, 10)
    kalkulator.tambah(2, 3, 5, 10)
    kalkulator.tambah(100)
    kalkulator.tambah()
    kalkulator.tambah(1, 2, "tiga", 4, 5.5)
```

```python
class DataProcessor:
  def __init__(self, processor_id):
    self.processor_id = processor_id
    print(f"Processor {self.processor_id} siap digunakan.")
  def process(self, *args, **kwargs):
    print(f"\n--- {self.processor_id} Memproses Data ---")
    if args:
      print("Argumen Diterima:")
      for i, arg in enumerate(args):
        print(f"args[{i}]: {arg} (tipe: {type(arg).__name__})")
    if kwargs:
      print("Keyword Arguments Diterima:")
    else:
      print("Tidak ada argumen posisi")

    if kwargs:
      print("Argumen Kata Kunci Diterima:")
      for kunci, nilai in kwargs.items():
        print(f"{kunci}: {nilai} (tipe: {type(nilai).__name__})")
    else:
      print("Tidak ada keyword arguments")
      print("-" * 20)

if __name__ == "__main__":
  processor1 = DataProcessor("DP-1")
  print("\nPanggilan 1: Tanpa Argumen Tambahan")
  processor1.process()

  print("\nPanggilan 2: Hanya Argumen Posisi")
  processor1.process(100, "Status OK", 99.9, False)

  print("\nPanggilan 3: Hanya Argumen Kata Kunci")
  processor1.process(user="admin", level = 5, mode = "verbose")

  print("\nPanggilan 4: Kombinasi Argumen Posisi dan Kata Kunci")
  processor1.process("Task=A", "Task=B", status="Running", priority="high", thread_id=54321)

  print("\nPanggilan 5: Argumen Posisi Kompleks")
  list_ids = ["A1", "B2", "C3"]
  dict_params = {"timeout":30, "retry":3}
  processor1.process(list_ids, dict_params, owner="System", enabled=True)
```

```python
import math
class Bentuk:
  def nama_bentuk(self):
    return "Bentuk Generik"
  def hitung_luas(self):
    raise NotImplementedError("Subclass harus mengimplementasikan metode ini")
class Lingkaran(Bentuk):
  def __init__(self, radius):
    self.radius = radius
  def nama_bentuk(self):
    return "Lingkaran"
  def hitung_luas(self):
    return math.pi * (self.radius ** 2)
class Persegi(Bentuk):
  def __init__(self, sisi):
    self.sisi = sisi
  def nama_bentuk(self):
    return "Persegi"
  def hitung_luas(self):
    return self.sisi * self.sisi
class TaplakMeja:
  def __init__(self, panjang, lebar):
    self.panjang = panjang
    self.lebar = lebar
  def hitung_luas(self):
    return self.panjang * self.lebar
  def deskripsi(self):
    return f"Taplak Meja {self.panjang}X{self.lebar}"
def tampilkan_info_luas(objek_dengan_luas):
  print("-" * 20)
  try:
    luas = objek_dengan_luas.hitung_luas()
    try:
      nama = objek_dengan_luas.nama_bentuk()
    except AttributeError:
      nama = type(objek_dengan_luas).__name__
    print(f"Objek: {nama}")
    print(f'Luas: {luas:.2f}')
  except AttributeError:
    print(f"Objek: {type(objek_dengan_luas).__name__} tiddak dapat dihitung luasnya (metode tidak ditemukan)")
  except NotImplementedError:
    print(f"Metode 'hitung luas' belum diimplementasikan untuk {type(objek_dengan_luas).__name__}")
if __name__ == "__main__":
  lingkaran1 = Lingkaran(7)
  persegi1 = Persegi(5)
  taplak1 = TaplakMeja(1.5, 0.8)
  bentuk_dasar = Bentuk()
  string_biasa = "Ini String"
  daftar_item = [lingkaran1, persegi1, taplak1, bentuk_dasar, string_biasa]
  print("Menampilkan Info Luas (Polimorfisme Campuran):")
  for item in daftar_item:
    tampilkan_info_luas(item)
```

```python
class Burung:
  def __init__(self, nama):
    self.nama = nama

  def terbang(self):
    print(f"{self.nama} terbang dengan cara umum.")

  def bersuara(self):
    print(f"{self.nama} mengeluarkan suara burung.")

class Elang(Burung):
  def __init__(self, nama, rentang_sayap):
    super().__init__(nama)
    self.rentang_sayap = rentang_sayap
  def terbang(self):
    print(f"{self.nama} terbang tinggi melayang diangkasa.")
  def bersuara(self):
    print(f"{self.nama} berteriak nyaring!")
  def berburu(self):
    print(f"{self.nama} mencari mangsa dari ketinggian.")

class Pipit(Burung):
  def __init__(self, nama, warna_bulu):
    super().__init__(nama)
    self.warna_bulu = warna_bulu
  def terbang(self):
    print(f"{self.nama} terbang cepat diantara pepohonan.")
  def bersuara(self):
    print(f"{self.nama} berkicau merdu: Cit Cit!")
  def membangun_sarang(self):
    print(f"{self.nama} sedang mengumpulkan ranting untuk sarang")

def interaksi_dengan_burung(burung):
  print(f"\n--- Berinteraksi dengan {type(burung).__name__}: {getattr(burung, 'nama', 'Objek Tidak Dikenal')} ---")
  if isinstance(burung, Burung):
    burung.terbang()
    burung.bersuara()
    if isinstance(burung, Elang):
      print("Ini adalah Elang")
      burung.berburu()
    elif isinstance(burung, Pipit):
      print("Ini adalah Pipit")
      burung.membangun_sarang()
    else:
      print("Ini adalah burung umum")
  else:
    print("Objek ini bukan burung.")
if __name__ == "__main__":
  elang_sumatra = Elang("Elang Sumatra", 1.8)
  pipit_rumah = Pipit("Pipit Rumah", "Abu-abu")
  merak = Burung("Merak")
  kucing = "Meong"

  daftar_makhluk = [elang_sumatra, pipit_rumah, merak, kucing]

  for makhluk in daftar_makhluk:
    interaksi_dengan_burung(makhluk)
```

```python
class Komputer:
  def __init__(self, merk, processor, ram_gb, tiperam):
    self.merk = merk
    self.tiperam = tiperam
    self.processor = processor
    self.ram_gb = ram_gb
    self.ram_ditambah = False

  def info_spesifikasi(self):
    print(f"Merk: {self.merk}")
    print(f"Processor: {self.processor}")
    print(f"RAM: {self.ram_gb} {self.tiperam}")

  def jalankan_app(self, nama_app):
    self.nama_app = nama_app
    print(f"Komputer {self.merk} sedang menjalankan aplikasi: {self.nama_app}")

  def tambah_ram(self, jumlah, tiperam):
    self.ram_gb += jumlah
    self.ram_ditambah = True

  def info_tambah_ram(self):
    if self.ram_ditambah:
      print(f"RAM komputer {self.merk} telah berhasil di-upgrade menjadi {self.ram_gb} {self.tiperam}")


class Laptop(Komputer):
  def __init__(self, merk, processor, ram_gb, ukuran_layar, berat, tiperam):
    super().__init__(merk, processor, ram_gb, tiperam)
    self.ukuran_layar = ukuran_layar
    self.berat = berat

  def info_spesifikasi(self):
    super().info_spesifikasi()
    print(f"Ukuran Layar: {self.ukuran_layar}")
    print(f"Berat: {self.berat} kg")


class Desktop(Komputer):
  def __init__(self, merk, processor, ram_gb, casing, monitor, tiperam):
    super().__init__(merk, processor, ram_gb, tiperam)
    self.casing = casing
    self.monitor = monitor

  def info_spesifikasi(self):
    super().info_spesifikasi()
    print(f"Casing: {self.casing}")
    print(f"Monitor: {self.monitor}")


def spek_komplit(komputer):
  print("--- Spesifikasi Komputer ---")
  komputer.info_spesifikasi()
  komputer.info_tambah_ram()
  print("-" * 40)


if __name__ == "__main__":
  laptop1 = Laptop("Asus", "I5", 16, "15.6 inch", 2, "DDR4")
  desktop1 = Desktop("Lenovo", "Ryzen 5", 32, "Bulat", "Samsung 4K", "DDR5")
  laptop2 = Laptop("HP", "I9", 8, "14.6 inch", 5, "DDR5")
  desktop2 = Desktop("Advan", "Ryzen 9", 48, "Tinggi", "Xiaomi 2K", "DDR4")

  daftar_komputer = [laptop1, laptop2, desktop1, desktop2]

  laptop1.tambah_ram(8, laptop1.tiperam)
  desktop2.tambah_ram(16, desktop2.tiperam)

  for komputer in daftar_komputer:
    spek_komplit(komputer)
```
